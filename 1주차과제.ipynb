# 1. 언어 분야 인공지능 서비스 사례


### 집단 지성 번역 플랫폼, ' 플리토 '
<br>

1-1. 플리토란?

<br>

- 2012년 서비스를 시작한 집단 지성 번역 관련 통합 플랫폼. 번역 중심 소셜 네트워크 서비스
- 초기에는 기계 번역을 사용하는 네이버 파파고, 구글 번역기와 같은 다른 번역체계와 다르게 이용자가 번역을 신청하면 번역가 회원이 직접 번역을 해주는 방식 -> 시스트란 인터네셔널과 협약하여 nmt기술 기반 번역 엔진을 도입, 인공지능 번역과 집단지성 번역을 함께 사용
- 인공지능 번역을 통해 자동으로 번역 된 내용에 더하여 다른 사용자의 번역과의 비교를 통해 이용자의 상황에 더 적합한 자연스러운 번역을 얻을 수 있게 함으로써 인공지능 번역의 한계를 극복 하고자 함
- 인공지능 번역의 성능을 결정하는 것은 올바른 학습 데이터 -> 집단지성을 이용하여 얻은 양질의 번역 빅데이터를 인공지능 번역기의 학습을 위한 데이터로써 필요로 하는 타 기업에 판매

<br>

1-2. 적용 기술

<br>

- 기존 기계번역 알고리즘과 한계

|알고리즘|특징|한계|
|:---|:---|:---|
|규칙 기반 기계번역<br>(Rule Based Machine Translation, PBMT)|번역 대상인 언어에 대해 높은 이해를 가진 개발자가<br>직접 정의한 규칙에 따라 번역하는 방법|- 번역기를 만들거나 새로운 언어를 추가할 때 마다<br>해당 언어에 능통한 개발 인력 필요<br> - 개발 및 유지보수에 비용과 인력 많이 필요<br>- 구어체나 문학적 표현이 들어간 경우 번역기의 효율 급감|
|통계 기반 기계 번역<br>(Statistical Machine Translation, SMT)|수많은 데이터로부터 자동으로 통계정보를 추출해<br>규칙을 정의하고 그 규칙에 따라 번역하는 방법|- 최소 200만 개 이상의 번역 데이터를 필요로 함<br>- 이를 관리하고 분석하기 위해 들어가는 시간과 비용도 매우 큼<br>- 전체 문장이 흘러가는 맥락에 대한 이해도 부족

  > 한계 극복을 위해 '인공신경망 번역(NMT)' 이라는 새로운 번역 기술 도입


- 인공신경망 번역(Neural Machine Translation, NMT)

  - RBMT와 SMT의 새로운 대안으로 제안된 AI가 적용된 번역 기술. 빅데이터로 학습된 인공신경망을 기반으로 함.
  - 신경망은 기계학습이 가능한 소프트웨어. 실제 뇌신경을 모방한 전산단위로 수백만 또는 수십억 개의 뉴런이 각각 입력한 정보를 다른 뉴런에세 전달해 지식을 확장시키는 방식이며, 복잡한 정보까지 학습할 수 있게 함.
  - 대용량의 학습 데이터와 몇 차원 벡터로 표현 할 것인지의 정보, 학습에 필요한 변수 정보만 지정해주면 기계가 알아서 번역하는 법을 학습함


- 인공신경망 번역(NMT)의 원리

  - SMT와 다르게 단어나 구 단위가 아닌 문장 전체를 대상으로 삼아 전체 문맥을 고려한 번역 결과를 제공함
  - 원문의 문장을 숫자 형태인 벡터값으로 치환(encoder) -> 이렇게 만들어진 문장 벡터를 기반으로 번역 결과를 얻고자 하는 대상언어의 문장을 생성(decoder)
  - 원문에 포함된 단어나 표현 등은 유사한 단어들과 관계를 맺고 있는 일정한 벡터값 가짐
  - 각 벡터값을 대상 언어와 조합해 최적의 가중치(weight parameter)를 갖는 문장 만듬

<br>

1-3. 인공지능 번역의 향후 전망

<br>

- NMT기술의 도입으로 인해 기존의 기계 번역보다 번역의 품질이 대폭 개선된 것은 사실이지만, 인간 번역가와 비교하기에는 아직 부족함이 많음. 문장 단위를 넘어서 전체 지문의 흐름과 문장 간의 관계, 비 언어적 정보, 문학 작품 번역과 같은 영역에서는 인간 번역을 따라갈 수 없음. 그러나 기계 번역은 인간 번역가가 따라갈 수 없는 속도로 번역이 가능하며, 사용 용도가 한정되어 있는 특정 분야에서는 충분히 경쟁력이 있음.


# 2. 음성 분야 인공지능 서비스 사례


### 전 세계 최초 음성인식 구동 스피커, ' 에코 '
<br>

2-1. 에코란?

<br>

- 아마존의 인공지능 비서 '알렉사'를 탑재한 AI스피커
- 알렉사는 모든 플랫폼과 기기에서 이용할 수 있는 '오픈 API형 인공지능 비서'로 머신러닝, 음성인식, 문장분석, 상황인지 등 인공지능 기술과 첨단 기숭릐 결합으로 사용자의 언어를 이해하고 사용자가 원하는 지시사항을 수행하는 소프트웨어 어플리케이션
- 기능
  1. 정보 검색 및 일정 관리 : 알렉사의 데이터베이스에 존재하는 엄청난 양의 정보에서 사용자의 질문에 적합한 정보를 찾아 대답 해 줌
  2. 음악 검색, 재생 기능 : 음성인식을 통해 사용자가 원하는 곡이나 동영상을 찾아 재생 해 줌
  3. 홈오토메이션 : 다양한 사물인터넷 도구를 음성 명령으로 제어 가능
  4. skill : 아마존 이외의 개발자들이 제공하는 확장팩 개념, 특정 업체의 기기나 인터넷 서비스를 알렉사와 연계해서 사용 할 수 있도록 해주는 기능
  5. 상품 주문 : 음성 명령으로 아마존에서 원하는 제품을 주문하는 기능

<br>

2-2. 적용 기술

<br>

- 인공지능 비서는 다른 인공지능들과 동일하게 머신러닝과 딥러닝을 활용. 각 IT업체의 클라우드 데이터 속에서 인공신경망을 적용한 인공지능 모델을 설치 한 후, 기계학습 기법을 통해 처리 능력과 처리 할 수 있는 분야를 확대시킨 것


- STT(Speech to Text)기술을 통해 사람의 음성을 텍스트로 옮김 -> 자연어 처리(NLP)기술을 통해 자연어를 분석하여 이해하고 처리 -> TTS(Text To Speech)기술을 통해 AI스피커가 사람의 말소리로 응답, 또는 요청한 기능 수행


- 자연어 처리(Natural Language Processing, NLP)

  - 인간의 언어인 자연어를 컴퓨터를 이용해 처리하기 위한 기술 분야
  - NLP의 핵심 기술은 형태소 분석 기술, 구문 분석 기술, 의미 분석 기술, 담화 분석 기술, 단어 및 문장 생성 기술이라고 할 수 있음
  - 자연어의 분석 단계
  > 자연언어문장 -> 형태소 분석 -> 구문 분석 -> 의미 분석 -> 화용 분석 -> 분석 결과

  - NPL에는 다양한 테크닉들이 쓰이고 있고, 수 많은 알고리즘과 개념들이 존재


- 워드 임베딩(Word Embedding) : 벡터화

  - 현재 NPL에서 가장 강력한 트렌드 중 하나로, 텍스트를 숫자로 표현하는 것
  - 단어를 사전을 사용하여 매핑하고 이를 벡터로 만드는 것
  > (ex) [the, cat, sat, on, mat]의 단여별로 구분하고 이를 0또는 1로 나타냄. 예를 들면 'cat'은 [0,1,0,0,0,0]이며, 'the'의 경우 [1,0,0,0,0]
  - BOW(Bag of words) : 주어진 말뭉치에서 사용된 각각의 단어에 인덱스를 매겨서 사전처럼 만들고 입력된 문서의 각 단어가 해당 단어 사전에서 해당 인덱스에 얼마나 나타나는지 표시하는 방식으로 입력된 문장을 분석하는 방법
  - Count Vector : 말뭉치에 여러개의 문서들이 있다고 했을 때 각 문서마다 고유의 토큰을 세어 이를 행렬로 나타내는 방법
  - TF-IDF Vector : 특정 단어가 문서 내에 얼마나 자주 등장하는지를 나타내는 TF(단어 빈도)와 어떤 단어가 문서 전체의 집합에서 얼마나 많이 나오는지를 나타내는 IDF(역문서 빈도)로 이 둘의 곱으로 값을 구함
  - word2vec : 단어를 벡터화 시켜 두 단어의 거리 유사성을 통해 의미적인 관계를 추론하는 방법, CBOW와 skip-gram의 두가지 방법이 있음
  - 이러한 여러 방식의 워드 임베딩을 통해 벡터화 된 데이터를 학습하고 예측하는 과정으로 자연어를 처리 할 수 있음

  <br>

2-3. 자연어 처리 기술의 향후 전망

  <br>

- 자연어 처리 기술은 인간과 컴퓨터의 소통이 필요한 산업 전반에 활용 가능. 자연어 처리에 대한 시장 수요도 급등하고 있음. 인공지능 연구의 발전을 위해서는 인간과 컴퓨터의 자유로운 소통은 반드시 이루어져야 하므로 자연어 처리에 대한 연구는 필수적임. 아직까지 인간 수준의 자연어 처리 능력을 부족해 보이지만 빠르게 발전 할 것으로 예상됨
  

# 3. 이미지 분야 인공지능 서비스 사례


### AI기반 이미지 검색 서비스, ' 스마트 렌즈 '
<br>

3-1. 스마트 렌즈란?

<br>

- 딥러님 기반의 이미지 분석 기술로 이름이나 키워드를 몰라도 이미지를 통해 관련 정보를 찾을 수 있는 이미지 검색 서비스
- 카메라를 물체에 가져다 대면 물체의 식별을 시도, 관련 검색 결과와 정보 표시
- 네이버의 스마트 렌즈, 구글의 구글 렌즈, 카카오 롱탭 이미지 검색 등이 있음


- 네이버 스마트 렌즈
  1. 스마트 렌즈 : 사진을 분석하여 유사한 이미지들을 검색
  2. 쇼핑 렌즈 : 네이버 쇼핑에서 유사한 상품과 쇼핑 정보를 찾아줌
  3. QR/바코드 렌즈 : QR/바코드를 인식해서 웹페이지나 상품정보를 제공
  4. 와인라벨 렌즈 : 촬영한 와인 라벨을 검색해서 평점과 랭킹, 자세한 정보 제공
  5. 상품라벨 렌즈 : 촬영한 상품의 라벨을 검색해서 평정과 자세한 정보 제공


- 구글의 구글 렌즈
  1. 구글 지도와 증강현실의 통합 : 구글 지도에서 카메라를 켜서 실제로 자신이 바라보는 현장에 대한 유용한 맥락 정보를 얻을 수 있음
  2. 지능형 텍스트 선택 : 카메라로 가리키는 것만으로 텍스트를 선택해 추출 할 수 있음
  3. 스타일 매치 : 카메라로 한 아이템을 가리키기만 해도 이에 어울리는 다른 아이템을 제안 해 줌
  4. 이미지 검색 : 카메라로 찾고 싶은 물건을 비추면 그와 관련된 정보를 얻을 수 있음
  5. 쇼핑 검색 : 구매 정보가 필요한 물건에 대해서 바코드로 검색 가능


- 카카오 롱탭 이미지 검색 : 채팅방의 사진을 길게 누른 뒤 '#검색'을 선택하면 유사한 이미지를 찾아줌

<br>

3-2. 적용 기술

<br>

- 구글 렌즈와 네이버 스마트렌즈 ,카카오의 롱탭 이미지 검색 모두 직접 알고리즘을 만드는 것이 아니라 데이터를 기반으로 모델을 만들어 문제를 해결하고자 하는 데이터 기반 방법(Data-driven approach)을 사용


- 수많은 이미지와 레이블이 있는 데이터셋을 통해 모델 학습을 하고, 이렇게 학습된 머신러닝 모델은 새로운 이미지를 인풋으로 받아 그 이미지의 레이블을 예측


- 네이버 스마트 렌즈, '스코픽(Scopic)'
  - 네이버 자체 비주얼 서치 기술
  - 네이버 스마트 렌즈의 차별성은 다른 서비스보다 이미지에 대한 키워드를 더 상세하게 제공한다는 것
  > (ex) '컵' 이나 '가구'같은 카테고리성 키워드 보다는 '식탁의자', '브라운 머그컵'같은 구체적인 키워드 제공
  - 딥러닝 기반의 이미지 분석 기술 + 이미지 자체의 정보와 주변 텍스트 정보를 함께 활용
  - 이미지에 태깅 학습 데이터를 만드는 작업이 생각보다 많은 비용과 시간이 들어가는데, 스코픽은 네이버가 보유한 사용자 데이터를 최대한 그대로 이용하는 방식을 취하는 것으로 충분히 많은 학습 데이터를 지속적으로 얻을 수 있음


- 구글 렌즈에 적용된 네가지 기술
  - 비전 AI와 구글의 지식 그래프를 기본 기술로 사용
  - 분류(Classification) : 이미지를 봤을 때 이미지 내 여러 개의 사물을 분류해내는 기술, 레이블링을 통해 사물에 정보를 부여
  - 발견(Detection) : 이름을 붙인 이미지의 각 사물의 위치를 파악하는 것
  - 임베딩(Embeddings) : 각 이미지를 머신 러닝으로 만들어 놓은 벡터값과 비교하여 수치화 하는 것
  - 모델 학습(Advances in model training) : 이미지 검색 속도를 빠르게 하기 위해서는 연산 속도가 중요함. 기존 GPU로 하던 머신 러닝을 커스텀 하드웨어인 TPU(Tensor Processing Unit)로 바꾸며 연산 속도의 개선을 이룸

<br>

3-3. 이미지 검색 기술의 향후 전망

<br>

- 이미지와 동영상은 텍스트보다 훨씬 많은 정보를 가지고 있기 때문에 더 폭넓게 활용 될 수 있음
- 쇼핑, 광고 등과 연계해 다양한 수익 모델 개발 가능성 존재
- 조금씩 차이가 존재하긴 하지만 이미지 검색 서비스의 기본적인 기능은 거의 비슷함. 그러므로 더 정확하고 활용도 있는 정보를 제공하는 것이 관건이기 때문에 양질의 학습 데이터를 확보하기 위한 노력 필요

# 4. 자율주행 분야 인공지능 서비스 사례


### 구글 웨이모(Waymo) 자율주행 택시 서비스, '웨이모 원(Waymo one)'
<br>

4-1. 웨이모 원(Waymo one)이란?

<br>

- 구글 모기업 알파벳의 자율주행차 부문인 웨이모(Waymo)의 자율주행 택시 서비스
- 스마트폰 앱으로 자율주행 택시를 호출하면 차량이 승객이 있는 지점까지 정확하게 태우고 목적지로 이동, 핸들이나 별도의 기기를 작동할 필요 없이 제한속도에 맞춰 스스로 운행하는 레벨4 자율주행차
- 아직 기술 완성도가 부족하여 안전성과 보안에 관련된 이슈가 완벽하게 해결되지 않음

<br>

4-2. 적용 기술

<br>

- 구글이 처음 자율주행 기술을 개발할 때, 수천 장의 사진을 분석하여 사진 속의 시각적 패턴을 식별하도록 컴퓨터를 가르칠 수 있는 알고리즘을 개발함, 그러나 이 알고리즘에 문제 발생 -> 전통적인 기계학습 방식을 신경망과 통합한 알고리즘 개발

- 자율주행 자동차는 센서를 통해 주행 중인 도로 상황을 인식하고, 딥 뉴럴 네트워크(Deep Neural Network, DNN)라고 불리는 알고리즘 배열을 사용하여 인식한 데이터를 처리하고 식별함


- 자율주행 성능을 위해서 중요한 것
  - 얼마나 많은 주행 데이터를 학습했는가
  - 학습한 데이터가 얼마나 고유한가 : 주행 데이터가 신경망 학습에 의미를 가져다 주기 위해서는 엣지 케이스, 무단횡단 상황과 같은 특이한 상황에 직면했을 경우 같은 흥미로운 데이터 필요


- PMT(Population-based training)
  - 구글의 인공지능 자회사 딥마인드가 개발한 비디오 게임 알고리즘 개선을 위한 기술
  - 알고리즘 모집단에서 최적표본(주어진 작업을 가장 효율적으로 수행하는 표본)에서 후보 코드를 채택, 특정 작업에 대한 기계 학습 알고리즘 및 매개 변수 선택 속도를 높이는 기술
  - 이 기술을 통해 자율주행 차량의 심화학습을 위한 처리 시간을 대폭 감소 할 수 있었고, 이를 통해 경쟁사들보다 빠른 기술 발전을 이룰 수 있었음

<br>

4-3. 자율주행 분야의 향후 전망

- 현재 웨이모 원은 4단계 자율주행 차량으로, 운행 하는 지역에도 제한이 있고 아직 한계점이 많음. 향후 완전 자율주행, 운전자와 운전대가 필요 없는 무인차를 선보인다는 것이 목표

- 자율주행 딥러닝 연구의 미래는 단순히 지각 능력이 있는 차량을 만드는 것을 넘어서 궁극적으로 기계학습 모델을 만드는 과정이 자동화되는 것
